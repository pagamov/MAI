## –ó–∞–¥–∞–Ω–∏–µ 3

```python
import os
import shutil

raw_dataset_path = '/content/drive/MyDrive/datasets/lab3'
raw_img_path = os.path.join(raw_dataset_path, 'video_0')
raw_lbl_path = os.path.join(raw_dataset_path, 'train.csv')

data_path = '/content/dataset/images'
lbl_path = '/content/dataset/labels'

os.mkdir('/content/dataset')
os.mkdir(data_path)
os.mkdir(lbl_path)

for img in os.listdir(raw_img_path):
  shutil.copy(
      os.path.join(raw_img_path, img),
      os.path.join(data_path, img))
```

–¢–µ–ø–µ—Ä—å –Ω—É–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å —Ñ–∞–π–ª—ã —Å —Ä–∞–∑–º–µ—Ç–∫–æ–π. –ß–∏—Ç–∞–µ–º —Ñ–∞–π–ª train.csv, —Ñ–∏–ª—å—Ç—Ä—É–µ–º –µ–≥–æ. –¢–∞–∫–∂–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –≤—Å–µ—Ö —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–π, —Ç–∞–∫ –∫–∞–∫ –∏–º–µ–Ω–Ω–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –æ–Ω–∏ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ train.csv.

```python
import pandas as pd

df = pd.read_csv(raw_lbl_path)
df = df[df['video_id'] == 0]

imgs = sorted(os.listdir(data_path), key = lambda x: int(x.split('.')[0]))
```

–í –¥–∞—Ç–∞—Å–µ—Ç–µ –æ—á–µ–Ω—å –º–Ω–æ–≥–æ –∫–∞—Ä—Ç–∏–Ω–æ–∫ –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –Ω–∏ –æ–¥–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ (–º–æ–∂–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∏–∂–µ). –î–ª—è —Ö–æ—Ä–æ—à–µ–≥–æ –æ–±—É—á–µ–Ω–∏—è –º—ã –≤–æ–∑—å–º–µ–º –ø–æ—á—Ç–∏ –≤—Å–µ —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–∏ —Å —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∏–º –æ–±—ä–µ–∫—Ç–æ–º –∏ –Ω–µ–º–Ω–æ–≥–æ –ø—É—Å—Ç—ã—Ö —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–π, —á—Ç–æ–±—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ—Ç—Ä–∞–±–∞—Ç—ã–≤–∞–ª—Å—è —Å–ª—É—á–∞–∏ —Å –ø—É—Å—Ç—ã–º–∏ –∫–æ—Ä—Ç–∏–Ω–∫–∞–º–∏ –ø—Ä–∏ —É–∂–µ —Ä–∞–±–æ—Ç–µ —Å –æ–±—É—á–µ–Ω–Ω–æ–π —Å–µ—Ç—å—é. –ü–æ—ç—Ç–æ–º—É –±—É–¥–µ–º –±—Ä–∞—Ç—å –ø—É—Å—Ç—ã–µ —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–∏ —Å–ª—É—á–∞–π–Ω–æ, –ø—Ä–∏—á–µ–º —Ç–∞–∫, —á—Ç–æ–±—ã –ø—Ä–æ—Ü–µ–Ω—Ç –ø—É—Å—Ç—ã—Ö —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–π –±—É–¥–µ—Ç —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ 5.

```python
print("Empty images:\t", (df['annotations'] == '[]').sum(), sep='\t')
print("Not empty images:", (df['annotations'] != '[]').sum(), sep='\t')

empty_imgs_ratio = 0.05

imgs_count = (df['annotations'] != '[]').sum() / (1 - empty_imgs_ratio)
empty_imgs_prob = (1 / (df['annotations'] == '[]').sum()) * imgs_count * empty_imgs_ratio
print("Empty images prob:", empty_imgs_prob, sep='\t')
```

–°–æ–∑–¥–∞–µ–º —Ñ–∞–π–ª—ã —Å –∞–Ω–æ—Ç–∞—Ü–∏—è–º–∏

```python
import random
from PIL import Image
import ast

img = Image.open(os.path.join(data_path, imgs[0]))
W = img.width
H = img.height

for annotation, img in zip(df['annotations'], imgs):
  if annotation != '[]':
    with open(os.path.join(lbl_path, f"{img.split('.')[0]}.txt"), 'w') as file:
      for location in ast.literal_eval(annotation):
        file.write(f"0 {location['x']/W:.6} {location['y']/H:.6} {location['width']/W:.6} {location['height']/H:.6}\n")
  else:
    if random.random() <= empty_imgs_prob:
      with open(os.path.join(lbl_path, f"{img.split('.')[0]}.txt"), 'w') as file:
        file.write('')
    else:
      os.remove(os.path.join(data_path, img))
```
