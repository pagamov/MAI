```python
import torch, torchvision
import numpy as np
from matplotlib import pyplot as plt
from torch.utils.data.dataloader import DataLoader
from tqdm import tqdm
```

```python
import cv2
import numpy as np
from torchvision.datasets import VOCSegmentation

cv2.setNumThreads(0)
cv2.ocl.setUseOpenCL(False)


VOC_CLASSES = [
    "background",
    "aeroplane",
    "bicycle",
    "bird",
    "boat",
    "bottle",
    "bus",
    "car",
    "cat",
    "chair",
    "cow",
    "diningtable",
    "dog",
    "horse",
    "motorbike",
    "person",
    "potted plant",
    "sheep",
    "sofa",
    "train",
    "tv/monitor",
]


VOC_COLORMAP = [
    [0, 0, 0],
    [128, 0, 0],
    [0, 128, 0],
    [128, 128, 0],
    [0, 0, 128],
    [128, 0, 128],
    [0, 128, 128],
    [128, 128, 128],
    [64, 0, 0],
    [192, 0, 0],
    [64, 128, 0],
    [192, 128, 0],
    [64, 0, 128],
    [192, 0, 128],
    [64, 128, 128],
    [192, 128, 128],
    [0, 64, 0],
    [128, 64, 0],
    [0, 192, 0],
    [128, 192, 0],
    [0, 64, 128],
]


class PascalVOCSearchDataset(VOCSegmentation):
    def __init__(self, root="~/data/pascal_voc", image_set="train", download=True, transform=None):
        super().__init__(root=root, image_set=image_set, download=download, transform=transform)

    @staticmethod
    def _convert_to_segmentation_mask(mask):
        # This function converts a mask from the Pascal VOC format to the format required by AutoAlbument.
        #
        # Pascal VOC uses an RGB image to encode the segmentation mask for that image. RGB values of a pixel
        # encode the pixel's class.
        #
        # AutoAlbument requires a segmentation mask to be a NumPy array with the shape [height, width, num_classes].
        # Each channel in this mask should encode values for a single class. Pixel in a mask channel should have
        # a value of 1.0 if the pixel of the image belongs to this class and 0.0 otherwise.
        height, width = mask.shape[:2]
        segmentation_mask = np.zeros((height, width, len(VOC_COLORMAP)), dtype=np.float32)
        for label_index, label in enumerate(VOC_COLORMAP):
            segmentation_mask[:, :, label_index] = np.all(mask == label, axis=-1).astype(float)
        return segmentation_mask

    def __getitem__(self, index):
        image = cv2.imread(self.images[index])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mask = cv2.imread(self.masks[index])
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
        mask = self._convert_to_segmentation_mask(mask)
        if self.transform is not None:
            transformed = self.transform(image=image, mask=mask)
            image = transformed["image"]
            mask = transformed["mask"]
        image,mask = torchvision.transforms.functional.to_tensor(image), torch.from_numpy(mask)
        mask = mask.permute(2,0,1)
        return image, mask
```

## Датасеты и аугментации
Для валидации просто загрузим валидационный датасет VOC. Однако для него будем использовать другие аугментации, точнее трансформацию, так как нам нужно, чтобы при каждой валидации были одни и те же значения, никакого рандома. Просто будем ресайзить картинку под наш входной формат.

```python
from albumentations import Compose, RandomBrightnessContrast, RandomCrop, Flip, RandomSizedCrop, OneOf, PadIfNeeded, Resize
```

```python
train_dataset = PascalVOCSearchDataset('.',download = False, image_set='train', transform = trainAugs)
test_dataset = PascalVOCSearchDataset('.',download = False, image_set='val', transform = testAugs)
```

## Метрики

Самая главная в данной ЛР метрика это IoU (Intersection over Union). Будем считать ее только для тех классов которые есть на картинке, для остальных же классов выставляем в качестве IoU nan. Алгоритм расчета был взят из этой статьи: https://medium.com/@cyborg.team.nitr/miou-calculation-4875f918f4cb

```python
from sklearn.metrics import confusion_matrix

def IoU(predict, lbl):
  lbl = lbl.detach().cpu()
  predict = predict.detach().cpu()

  lbl = lbl.argmax(1)
  predict = predict.softmax(1).argmax(1)

  lbl_flat = np.reshape(lbl, (len(lbl), 256*256))
  predict_flat = np.reshape(predict, (len(predict), 256*256))

  def bc_n(a):
    a = np.bincount(a)
    return np.pad(a, (0, 21 - a.shape[0]), 'constant', constant_values=0)

  actual_count = np.apply_along_axis(bc_n, 1, lbl_flat)
  pred_count = np.apply_along_axis(bc_n, 1, predict_flat)


  I = np.zeros(shape=(len(predict), len(VOC_CLASSES)))
  for i in range(len(predict)):
    c_matrix = confusion_matrix(lbl_flat[i], predict_flat[i], labels=np.arange(21))
    I[i] = np.diag(c_matrix)

  U = actual_count + pred_count * actual_count.astype(bool) - I
  not_zero_inds = np.nonzero(U)

  out = np.empty(shape=I.shape)
  out[:] = np.nan
  out[not_zero_inds] = I[not_zero_inds] / U[not_zero_inds]
  return out
```

Здесь же мы считаем разные MIoU. MIoU - это средняя IoU, Но вот средняя по чему?)

**MIoU** - Средняя IoU по всему датасету

**MIoU_over_classes** - Средняя IoU для каждого класса

**MIoU_over_images** - Средняя IoU для каждой картинки 

```python
def MIoU(i_o_u_matrix):
  return np.nanmean(i_o_u_matrix.flatten())
```

```python
def MIoU_over_classes(i_o_u_matrix):
  return np.apply_along_axis(np.nanmean, 0, i_o_u_matrix)
```

```python
def MIoU_over_images(i_o_u_matrix):
  return np.apply_along_axis(np.nanmean, 1, i_o_u_matrix)
```

Делаем модель. Спойлер. Почти никаких изменений делать не пришлось. Отличие разве что в том что даталоудера два: train и test.

```python
from torchvision.models.segmentation import deeplabv3_mobilenet_v3_large
```

```python
model = deeplabv3_mobilenet_v3_large(True, num_classes = len(VOC_CLASSES), progress = True)
```

```python
optim = torch.optim.AdamW(model.parameters(), lr = 1e-3, weight_decay = 1e-2)
```

```python

trainDataloader = DataLoader(train_dataset, batch_size = 8, shuffle = True, num_workers = 2)
testDataloader = DataLoader(test_dataset, batch_size = 8, num_workers = 2)
```

```python
crit = torch.nn.CrossEntropyLoss()
```

```python
model.cuda()
```

## Основной цикл обучения
Здесь:
1. Добавлен цикл по валидационным данным. Используем model.eval() и torch.no_grad() в качетве хорошего тона) 
2. Добавлен расчет таблицы IoU для train и test проходов
3. Изменен вывод. Теперь выводим loss и MIoU посчитаный по таблице IoU для train и test

```python
r_losses_train = []
r_losses_test = []
next_desc = "Epoch 0"
for epoch in range(1):
  rlossTrain = 0
  i_o_u_matrixTrain = np.zeros(shape=(len(train_dataset), len(VOC_CLASSES)))
  model.train()
  for i, (img,lbl) in enumerate(tqdm(trainDataloader, desc = next_desc)):
    batch_size_l = len(img)
    img = img.cuda()
    lbl = lbl.cuda()
    predict = model(img)['out']
    loss = crit(predict, lbl)
    loss.backward()
    optim.step()
    optim.zero_grad()
    rlossTrain += loss.detach().cpu().item() / len(trainDataloader)
    i_o_u_matrixTrain[i*batch_size_l:i*batch_size_l+batch_size_l] = IoU(predict, lbl)

  rlossTest = 0
  i_o_u_matrixTest = np.zeros(shape=(len(test_dataset), len(VOC_CLASSES)))
  model.eval()
  with torch.no_grad():
    for i, (img,lbl) in enumerate(tqdm(testDataloader)):
      batch_size_l = len(img)
      img = img.cuda()
      lbl = lbl.cuda()
      predict = model(img)['out']
      loss = crit(predict, lbl)
      rlossTest += loss.detach().cpu().item() / len(trainDataloader)
      i_o_u_matrixTest[i*batch_size_l:i*batch_size_l+batch_size_l] = IoU(predict, lbl)
  
  r_losses_train += [rlossTrain]
  r_losses_test += [rlossTest]
  miou_train = MIoU(i_o_u_matrixTrain)
  miou_test = MIoU(i_o_u_matrixTest)
  print(f"Training scores. loss: {rlossTrain:.3f},\tMIoU: {miou_train:.3f}")
  print(f"Testing scores.  loss: {rlossTest:.3f},\tMIoU: {miou_test:.3f}")
  next_desc = f"Epoch {epoch+1}"
  
```

# Пример картинок и их IoU векторов

Функция работы модели на заданных индексах из валидационного датасета

```python
from math import ceil
def testModelOn(idxs):
    img = []
    lbl = []
    for i in idxs:
        im,  lb = test_dataset[i]
        img.append(im)
        lbl.append(lb)

    img = torch.stack(img)
    lbl = torch.stack(lbl)

    img = img.cuda()
    lbl = lbl.cuda()
    predict = model(img)['out']

    fig,axes = plt.subplots(len(lbl), 3, figsize = (12, len(lbl) * 4))

    print(IoU(predict, lbl))

    print([np.nanmean(IoU(predict, lbl)[i]) for i in range(len(lbl))])

    for idc, (simg, spred, slbl) in enumerate(zip(img.detach().cpu(), predict.detach().cpu().softmax(1), lbl.detach().cpu())):
      axes[idc, 0].imshow(simg.permute(1,2,0))
      axes[idc, 1].imshow(spred.argmax(0), vmin = 0, vmax = len(VOC_CLASSES)-1)
      axes[idc, 2].imshow(slbl.argmax(0), vmin = 0, vmax = len(VOC_CLASSES)-1)
    [ax.get_xaxis().set_visible(False) for ax in fig.axes]
    [ax.get_yaxis().set_visible(False) for ax in fig.axes]
    fig.show()
```

Считаем индексы для заданных перцентилей

```python
images_miou = MIoU_over_images(i_o_u_matrixTest)

sorted_images_miou = sorted(enumerate(images_miou), key=lambda x: x[1] if x[1] != 1 else 0)

best10 = [sorted_images_miou[ceil(len(sorted_images_miou)*x)][0] for x in [0.98, 0.99]]
best1525 = [sorted_images_miou[ceil(len(sorted_images_miou)*x)][0] for x in [0.77, 0.82]]
best4555 = [sorted_images_miou[ceil(len(sorted_images_miou)*x)][0] for x in [0.46, 0.54]]
```

Смотрим на картинки из заданных перцентилей

```python
testModelOn(best10)
```

```python
testModelOn(best1525)
```

```python
testModelOn(best4555)
```

Берем картинки у которых самое большое  IoU по зпдпнным классам.

```python
plane_mious = i_o_u_matrixTest[:,1]
bus_mious = i_o_u_matrixTest[:,6]
cow_mious = i_o_u_matrixTest[:,10]

best_im_in_clesses = [sorted(enumerate(mious), key=lambda x: x[1] if x[1] != 1 else 0)[-1][0] for mious in [plane_mious, bus_mious, cow_mious]]

testModelOn(best_im_in_clesses)
```
