## Reshape of Array

__Reshape__  - срабатывает, если итоговое число элементов будет равно исходному числу элементов. Где это возможно, данный мето

```python
grid = np.arange(1,10).reshape((3,3))
print(grid)


print(grid.reshape(1,9))

print(grid)
```

#### To row vector

# Array Concatenation

```python
x = np.array([1,2,3])
y = np.array([3,2,1])
print("concatenate array by row(axis = 0):",np.concatenate([x,y],axis = 0))
```

```python
grid = np.array([[1,2,3],[4,5,6]])
grid
```

```python
# for manydimensional array by axis 0 we add rows to the end 
print("concatenate array by row(axis = 0):",np.concatenate([grid,grid],axis = 0))
# for manydimensional array by axis 0 we add columns to the end 
print("concatenate array by row(axis = 1):",np.concatenate([grid,grid],axis = 1))
```

```python
print("concatenation with np.c_",np.c_[grid,grid])
print("concatenation with np.r_",np.r_[grid,grid])
```

```python
x = np.array([1,2,3])
grid = np.array([[9,8,7],[6,5,4]])
print("concatenate array by row(axis = 0):",np.vstack([x,grid]),sep = "\n")
print("concatenate array by row(axis = 0):",np.vstack([grid,x]),sep = "\n")

y = np.array([[99],[99]])
print("concatenate array by row(axis = 1):",np.hstack([y,grid]),sep = "\n")
print("concatenate array by row(axis = 1):",np.hstack([grid,y]),sep = "\n")
```

#### Splitting of arrays 

```python
x = [1,2,3,99,99,3,2,1]
x1,x2,x3 = np.split(x,[3,5])
print(x1,x2,x3)
```

```python
grid = np.arange(16).reshape((4,4))
grid
```

```python
upper,lower = np.vsplit(grid,[2])
print(upper)
print(lower)
```

```python
left,right = np.hsplit(grid,[2])
print(left)
print(right)
```

# KNN

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
X = np.random.random((10,2))
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
a = np.eye(2)
a[:,np.newaxis,:] - a[np.newaxis,:,:]
```

```python
differences = np.sum((X[:,np.newaxis,:] - X[np.newaxis,:,:])**2,axis = -1)
```

```python
differences.shape
```

```python
differences.diagonal()
```

```python
nearest = np.argsort(differences,axis = 1)
nearest
```

```python
 K = 2
 nearest_partition = np.argpartition(nearest, K + 1,axis = 1)
```

```python
nearest_partition
```

```python
plt.scatter(X[:,0],X[:,1],s = 100)
for idx in range(X.shape[0]):
  for jdx in nearest_partition[idx, :K+1]:
    plt.plot(*zip(X[jdx],X[idx]),color='black')
```
